```python
def get_most_recent_file_date():
    import os
    import glob
    import re

    # Directory where your .txt files are located
    directory = 'C:\\Users\\juanc\\Downloads'

    # Find all .txt files that match the pattern
    file_pattern = os.path.join(directory, "ProQuestDocuments-*.txt")
    files = glob.glob(file_pattern)

    if files:
        # Get the most recent file based on the last modified time
        most_recent_file = max(files, key=os.path.getmtime)
        
        # Open and read the most recent file
        with open(most_recent_file, 'r', encoding='utf-8') as file:
            content = file.read()
            
        anio_publicacion_match = re.search(r'Fecha de publicación: (.+)', content)
        if anio_publicacion_match:
            fecha = anio_publicacion_match.group(1).strip()
        # parse date 
        from datetime import datetime
        fecha2 = datetime.strptime(fecha, '%b %d, %Y')
        return fecha2
    else:
        print("No files matching the pattern were found.")
```
